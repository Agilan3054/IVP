```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Butterfly.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
image_blurred = cv2.GaussianBlur(image, (5, 5), 0)

roberts = cv2.filter2D(image_blurred, -1, np.array([[1, 0], [0, -1]]))
roberts = np.abs(roberts)

sobelx = cv2.Sobel(image_blurred, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(image_blurred, cv2.CV_64F, 0, 1, ksize=5)
sobel = np.sqrt(sobelx**2 + sobely**2)

prewittx = cv2.filter2D(image_blurred, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]))
prewitty = cv2.filter2D(image_blurred, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt = np.sqrt(prewittx**2 + prewitty**2)

laplacian = cv2.Laplacian(image_blurred, cv2.CV_64F, ksize=5)

canny = cv2.Canny(image_blurred, 50, 150)

plt.figure(figsize=(12, 10))

plt.subplot(2, 3, 1)
plt.imshow(image_blurred, cmap='gray')
plt.title('Original')

plt.subplot(2, 3, 2)
plt.imshow(roberts, cmap='gray')
plt.title('Roberts')

plt.subplot(2, 3, 3)
plt.imshow(sobel, cmap='gray')
plt.title('Sobel')

plt.subplot(2, 3, 4)
plt.imshow(prewitt, cmap='gray')
plt.title('Prewitt')

plt.subplot(2, 3, 5)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')

plt.subplot(2, 3, 6)
plt.imshow(canny, cmap='gray')
plt.title('Canny')

plt.show()
```
