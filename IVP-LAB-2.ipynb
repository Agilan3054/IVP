```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("spider.jpeg")
```

# Resizing and Interpolation operations

```python
plt.imshow(image)
```

```python
 
half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610))
 
stretch_near = cv2.resize(image, (780, 540), 
               interpolation = cv2.INTER_LINEAR)
 
 
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```

```python
new_height,new_width = 96,96
img = cv2.resize(image, dsize=(new_height, new_width), 
      interpolation=cv2.INTER_NEAREST)
plt.imshow(img)
```

```python
img = cv2.resize(image, dsize=(new_height, new_width), 
      interpolation=cv2.INTER_LINEAR)
plt.imshow(img)
```

```python
img = cv2.resize(image, dsize=(new_height, new_width), 
      interpolation=cv2.INTER_LINEAR_EXACT)
plt.imshow(img)
```

```python
img = cv2.resize(image, dsize=(new_height, new_width), 
      interpolation=cv2.INTER_LANCZOS4)
plt.imshow(img)
```

# Type conversions

```python
import cv2
import matplotlib.pyplot as plt

def convert_color_spaces(img):
    hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsl_image = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    grayscale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    return hsv_image, hsl_image, grayscale_image


img = image


hsv_image, hsl_image, grayscale_image = convert_color_spaces(img)

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title('BGR Image')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))
axes[1].set_title('HSV Image')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(hsl_image, cv2.COLOR_HLS2RGB))
axes[2].set_title('HSL Image')
axes[2].axis('off')


axes[3].imshow(grayscale_image, cmap='gray')
axes[3].set_title('Grayscale Image')
axes[3].axis('off')

plt.show()
```

```python
cv2.imwrite('output_image.jpg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])
cv2.imwrite('output_image.png', img)
cv2.imwrite('output_image_hsv.jpg', hsv_image, [int(cv2.IMWRITE_JPEG_QUALITY), 95])
cv2.imwrite('output_image_hsv.png', hsv_image)
```

#  Extraction of color bands from an image

```python
img = cv2.imread("india.jpeg")
img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img = img_bgr

b, g, r = cv2.split(img)
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(b.ravel(), bins=256, color='blue', alpha=0.5, label='Blue', rwidth=0.8)
ax.hist(g.ravel(), bins=256, color='green', alpha=0.5, label='Green', rwidth=0.8)
ax.hist(r.ravel(), bins=256, color='red', alpha=0.5, label='Red', rwidth=0.8)
ax.set_title('Color Distribution in Channels')
ax.set_xlabel('Pixel Intensity')
ax.set_ylabel('Frequency')

ax.legend()

plt.show()
```
