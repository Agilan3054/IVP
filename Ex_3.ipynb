```python
import cv2
import numpy as np
from skimage.util import random_noise
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
uniform_noise = np.zeros_like(image, dtype=np.uint8)
cv2.randu(uniform_noise, 0, 255)
noisy_image_uniform = cv2.add(image, uniform_noise)
plt.imshow(noisy_image_uniform)
```

```python
gaussian_noise = np.zeros_like(image, dtype=np.uint8)
cv2.randn(gaussian_noise, 0, 255)
noisy_image_gaussian = cv2.add(image, gaussian_noise)
plt.imshow(noisy_image_gaussian)
```

```python
noisy_image_salt_pepper = random_noise(image, mode='s&p', amount=0.5)
plt.imshow(noisy_image_salt_pepper)
```

```python
blurred_image = cv2.blur(image,
                         (5, 5))
plt.imshow(blurred_image)
```

```python
blurred_gaussian_image = cv2.GaussianBlur(image, (5, 5), 0)
plt.imshow(blurred_gaussian_image)
```

```python
blurred_median_image = cv2.medianBlur(image, 5)
plt.imshow(blurred_median_image)
```

```python
image_night = cv2.imread('NightVision.jpg')

gray_image = cv2.cvtColor(image_night, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
contrasted_image = cv2.convertScaleAbs(gray_image, alpha=1.5, beta=2)
plt.imshow(contrasted_image, cmap='gray')
```

```python
plt.hist(gray_image.ravel(), bins=256)
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Histogram of Image')
plt.show()
```

```python
equalized_image = cv2.equalizeHist(gray_image)
plt.imshow(equalized_image, cmap='gray')
```

```python
plt.hist(equalized_image.ravel(), bins=256)
plt.show()
```

```python
from skimage.exposure import match_histograms

matched = match_histograms(image_night, image)
plt.imshow(matched)
```
